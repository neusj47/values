```python
import finterstellar as fs
OTP = ''
```

```python
df_BIO = fs.fn_single(otp = OTP, symbol = 'BIO', window = 'T')
```

```python
df_BIO['PER'] = df_BIO['Price'] / df_BIO['EPS']
df_BIO['PBR'] = df_BIO['Price'] / (df_BIO['Shareholders Equity'] / df_BIO['Shares'])
```

```python
fs.draw_chart(df_BIO, left = 'Price', right = 'PER')
```

```python
fs.draw_chart(df_BIO, left = 'Price', right = 'PBR')
```

```python
df_BIO['NCAV'] = df_BIO['Current Assets'] - df_BIO['Total Liabilities']
df_BIO['Market Cap'] = df_BIO['Price'] - df_BIO['Shares']
df_BIO['Safety Margin'] = df_BIO['NCAV'] - df_BIO['Market Cap']*1.5
```

```python
fs.draw_chart(df_BIO, left = 'Price', right = 'Safety Margin')
```

```python
# 1. 안전마진 계산 
terms = fs.set_terms(trade_start='2019Q3', trade_end='2021Q3')
data = {}
for t in terms:
    data[t] = fs.fn_consolidated(otp=OTP, term=t, vol=10000)
    data[t]['NCAV'] = data[t]['Current Assets'] - data[t]['Total Liabilities']
    data[t]['Market Cap'] = data[t]['Price_M3'] * data[t]['Shares']
    data[t]['Safety Margin'] = data[t]['NCAV'] - data[t]['Market Cap']*1.5
```

```python
s = {}
signal = {}
for t in terms:
    s[t] = fs.fn_filter(data[t], by='Safety Margin', floor=0, n=1, asc=False)
    signal[t] = (s[t].index)
```

```python
signal
```

```python
df = fs.backtest(signal=signal, data=data, m=3, cost=.001)
```

```python
fs.draw_return(df)
```
